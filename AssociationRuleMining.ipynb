```python
from itertools import combinations

transactions = [
    {'Pain', 'Beurre', 'sucre'},
    {'lait', 'sucre', 'Beurre'},
    {'Pain', 'Beurre', 'sucre', 'Mayo'},
    {'Beurre', 'Pain'}
]


min_support = 2


def get_support(itemset, transactions):
    return sum(1 for t in transactions if itemset.issubset(t))


def apriori(transactions, min_support):

    items = sorted({item for t in transactions for item in t})


    L = []
    L1 = [{i} for i in items if get_support({i}, transactions) >= min_support]
    L.append(L1)

    k = 1
    while L[k - 1]:
       
        candidates = []
        prev_L = L[k - 1]
        items_prev = [sorted(list(x)) for x in prev_L]

        for i in range(len(items_prev)):
            for j in range(i + 1, len(items_prev)):
                
                union = sorted(set(items_prev[i]) | set(items_prev[j]))
                if len(union) == k + 1 and set(union) not in candidates:
                    candidates.append(set(union))

       
        Lk = [c for c in candidates if get_support(c, transactions) >= min_support]
        L.append(Lk)
        k += 1

    return L



frequent_itemsets = apriori(transactions, min_support)


for i, Lk in enumerate(frequent_itemsets):
    if Lk:
        print(f"\nFrequent {i+1}-itemsets:")
        for itemset in Lk:
            print(f"{itemset} - support = {get_support(itemset, transactions)}")
```
